```python
snapshot_date = df['InvoiceDate'].max()
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

rfm.columns = ['Recency','Frequency','Monetary']

rfm.head()
```

```python
rfm['R_score'] = pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method="first"),4,labels=[1,2,3,4])
rfm['M_score'] = pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])
```

```python
rfm['RFM_Score'] = rfm[['R_score','F_score','M_score']].astype(str).sum(axis=1)
rfm.head()
```

```python
customer_revenue = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)

customer_revenue.head()
```

```python
pareto = customer_revenue.reset_index()

pareto['cum_revenue'] = pareto['TotalPrice'].cumsum()
pareto['cum_percentage'] = 100 * pareto['cum_revenue'] / pareto['TotalPrice'].sum()

pareto.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.bar(range(len(pareto)), pareto['TotalPrice'])

plt.plot(range(len(pareto)), pareto['cum_percentage'], color='red')

plt.axhline(80, color='green', linestyle='--')

plt.title("Pareto Analysis of Customer Revenue")

plt.xlabel("Customers (sorted by revenue)")
plt.ylabel("Revenue / Cumulative %")
plt.savefig("../outputs/pareto_analysis.png", dpi=300)

plt.show()
```

```python
pareto_80 = pareto[pareto['cum_percentage'] <= 80]

num_customers = len(pareto_80)
total_customers = pareto['CustomerID'].nunique()

print("Customers contributing to 80% revenue:", num_customers)
print("Percentage of customers:", (num_customers/total_customers)*100)
```
